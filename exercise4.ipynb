```python
import pandas as pd
import matplotlib.pyplot as plt

data_path = '/Users/sylvia/Documents/HIST-1354-YidiWang/census_data.csv'
df = pd.read_csv(data_path)

# Filter the data for Davidson County
davidson_data = df[df['county'] == 370] 

# Calculate total population for each year
ppl_1820 = davidson_data[davidson_data['year'] == 1820]['citypop'].sum()
ppl_1830 = davidson_data[davidson_data['year'] == 1830]['citypop'].sum()
ppl_1840 = davidson_data[davidson_data['year'] == 1840]['citypop'].sum()

# Create a DataFrame for display
population_trend = pd.DataFrame({
    'Year': [1820, 1830, 1840],
    'Population': [ppl_1820, ppl_1830, ppl_1840]
})

# Displays full numbers instead of le6
plt.ticklabel_format(style='plain', axis='y')

# Create a line plot for Davidson County Population
plt.plot(population_trend['Year'], population_trend['Population'], marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Davidson County Population Trend (1820-1840)')
plt.grid(True)
plt.show()
```
